# Summary

Let's recap a few key points:

- Use `nvcc` to compile CUDA C++
- CUDA programs are never GPU-only
- CUDA programs always start on CPU
- There are no "magical" execution space changes
- Add `__host__ __device__` to functions that can be executed on GPU
- All the useful algorithms in `std::` are available in `thrust::`!
